```python
import numpy as np  
import matplotlib.pyplot as plt  
from sklearn.linear_model import LinearRegression  
```

```python
# Input (Study Hours) and Output (Exam Scores)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Reshape for sklearn  
y = np.array([50, 55, 60, 70, 75])  
```

```python
model = LinearRegression()  
model.fit(X, y)  
```

```python
predicted_score = model.predict([[6]])  
print(f"Predicted Score for 6 hours: {predicted_score[0]:.2f}")  
```

```python
# Plot data points
plt.scatter(X, y, color='blue', label='Actual Data')

# Plot regression line
plt.plot(X, model.predict(X), color='red', label='Regression Line')

# Predict for new values (including 6 hours)
plt.scatter([[6]], predicted_score, color='green', marker='x', s=100, label='Predicted Score')

plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.legend()
plt.show()
```
