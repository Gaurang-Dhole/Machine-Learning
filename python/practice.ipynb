```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
x = np.array([0, 5, 10, 15, 20, 25])
y = np.array([0,25,100,225,400,625])
```

```python
x.shape
```

```python
plt.scatter(x, y, color='orange', label='Data Points')
```

# Liner Regression


```python
x = x.reshape(-1, 1)
model = LinearRegression() 
model.fit(x, y)
predicted_score = model.predict([[6]])  
predicted_score
```

```python
plt.scatter(x, y, color='orange', label='Data Points')
plt.plot(x, model.predict(x), color='red', label='Regression Line')
plt.scatter([[6]], predicted_score, color='green', marker='x', s=100, label='Predicted Score')
```

# Polynomial Regression


```python
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)
x_p = poly.fit_transform([[6]])
```

```python
model = LinearRegression()
model.fit(x_poly, y)
y_pred = model.predict(x_poly)
predicted = model.predict(x_p)
predicted
```

```python
plt.scatter(x, y, color='orange', label='Actual Data')  
plt.plot(x, y_pred, color='red', label='Polynomial Regression Curve')  
plt.scatter([[6]], predicted, color='green', marker='+', s=100, label='Predicted Score')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Polynomial Regression (Degree=2)')
plt.legend()
plt.show()
```
