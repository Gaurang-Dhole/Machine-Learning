```python
X = db["YearsExperience"].values
y = db["Salary"].values
```

```python
m, n = db.shape
print(f"{m}, {n}")
```

```python
db = db.sample(frac=1,random_state=42).reset_index(drop=True)
```

```python
split_index = int(0.7*len(db))
train_data = db[:split_index]
test_data =  db[split_index:]
```

```python
X_train = train_data["YearsExperience"].values
y_train = train_data["Salary"].values

X_test = test_data["YearsExperience"].values
y_test = test_data["Salary"].values
```

```python
plt.scatter(X,y)
```

```python
plt.scatter(X_train,y_train)
```

```python
plt.scatter(X_test,y_test)
```

```python
m = 0
b = 0
alpha = 0.00001
it = 100000
n = len(X_train)

for it in range(it):
    y_pred = m* X_train + b
    error = y_train - y_pred


    dm = (-2/n)*sum(X_train* error)
    db = (-2/n)*sum(error)


    m = m - alpha*dm
    b = b - alpha*db

    if it % 100 == 0:
        mse = (error**2).mean()
        print(f"Epoch{it},MSE: {mse},m:{m},b:{b}")
```

```python
y_pred = m * X_test + b
```

```python
from sklearn.metrics import r2_score

print(f"R2: {r2_score(y_test, y_pred)}")
```

```python
mse = np.mean((y_test - y_pred)**2)
print(f"\nFinal MSE on Test Data: {mse:.2f}")
```

```python
plt.scatter(X_test, y_test, color='red', label="Actual")
plt.plot(X_test, y_pred, color='blue', label="Predicted Line")
plt.xlabel("YearsExperience (normalized)")
plt.ylabel("Salary")
plt.title("Linear Regression from Scratch")
plt.legend()
plt.show()
```

```python
y_train_pred = m * X_train + b

plt.figure(figsize=(8,6))
plt.scatter(X_train, y_train, color='red', label='Actual Data')
plt.plot(X_train, y_train_pred, color='blue', label='Regression Line', linewidth=2)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Linear Regression Line (All Data)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
y_full_pred = m * X + b

plt.figure(figsize=(8,6))
plt.scatter(X, y, color='red', label='Actual Data')
plt.plot(X, y_full_pred, color='blue', label='Regression Line', linewidth=2)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Linear Regression Line (All Data)")
plt.legend()
plt.grid(True)
plt.show()
```
