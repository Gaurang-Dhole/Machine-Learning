```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
db = pd.read_csv("Salary.csv")
```

```python
db.head()
```

```python
db.describe()
```

```python
db.isnull()
```

```python
db.isnull().sum()
```

```python
X = db[["YearsExperience"]]
y = db[["Salary"]]
```

```python
X
```

```python
y
```

```python
m, n = db.shape
print(f"{m}, {n}")
```

```python
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size=0.3, random_state=56)
```

```python
plt.scatter(X,y)
```

```python
plt.scatter(X_train,y_train)
```

```python
plt.scatter(X_test,y_test)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
print(f"R2: {r2_score(y_test, y_pred)}")
```

```python
plt.scatter(X_test, y_test, color='red', label="Actual")
plt.plot(X_test, y_pred, color='blue', label="Predicted Line")
plt.xlabel("YearsExperience (normalized)")
plt.ylabel("Salary")
plt.title("Linear Regression from Scratch")
plt.legend()
plt.show()
```

```python
c = lr.intercept_
m = lr.coef_[0]
y_pred_full = m * X + c 
plt.scatter(db['YearsExperience'], db['Salary'])
plt.plot(X, y_pred_full, color='red')
```
