```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_digits
```

```python
data = load_digits()
```

```python
data.data.shape
```

```python
X = data.data
y = data.target
```

```python
import matplotlib.pyplot as plt
```

```python
plt.matshow(data.images[9], cmap="gray")
y[9]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=12)
X_test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

model = LogisticRegression(multi_class='ovr', n_jobs=-1, max_iter=1000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
print(classification_report(y_test, y_pred))
```

```python
sample_data = [
    0,   0,   0,   0,  20,  20,   0,   0,
    0,   0,   0,  25,  20,  20,   0,   0,
    0,   0,  25,  25,  25,  20,   0,   0,
    0,   0,   0,   0,  25,  20,   0,   0,
    0,   0,   0,   0,  25,  20,   0,   0,
    0,   0,   0,   0,  25,  15,   0,   0,
    0,   0,   0,   0,  25,  15,   0,   0,
    0,   0,   0,   0,  20,  20,   0,   0
]

sample_data_2d = np.array(sample_data).reshape(8, 8)

plt.matshow(sample_data_2d, cmap="gray")

sample_data_for_model = sample_data_2d.reshape(1, -1)

y_prod = model.predict(sample_data_for_model)

print(f"Predicted digit: {y_prod}")
```
